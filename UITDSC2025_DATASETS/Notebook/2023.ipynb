## Chỉ số PAPI tổng hợp (không có trọng số)

```python
column_name = 'Chỉ số PAPI tổng hợp (không có trọng số)'

# Tùy chỉnh giao diện
plt.figure(figsize=(10, 6))  # Kích thước biểu đồ

# Vẽ histogram
sns.barplot(data=df, y='Tỉnh/Thành phố', x=column_name, color='skyblue', orient='h')

# Thêm tiêu đề và nhãn
plt.title('Chỉ số PAPI theo Tỉnh/Thành phố', fontsize=14)
plt.xlabel('Chỉ số PAPI tổng hợp (không có trọng số)', fontsize=12)
plt.ylabel('Tỉnh/Thành phố', fontsize=12)

# Thêm lưới để dễ quan sát
plt.grid(True, axis='x', alpha=0.3)

# Hiển thị biểu đồ
plt.show()

# Tính và hiển thị các thống kê cơ bản
mean_val = df[column_name].mean()
median_val = df[column_name].median()
std_val = df[column_name].std()
min_val = df[column_name].min()
max_val = df[column_name].max()

print(f"Mean: {mean_val:.2f}")
print(f"Median: {median_val:.2f}")
print(f"Standard Deviation: {std_val:.2f}")
print(f"Minimum Value: {min_val:.2f}")
print(f"Maximum Value: {max_val:.2f}")

# Hiển thị biểu đồ
plt.show()
```

```python
# Vẽ histogram
sns.histplot(data=df, x=column_name, bins=10, kde=True, color='skyblue')

# Thêm tiêu đề và nhãn
plt.title(f'Histogram of {column_name}', fontsize=14)
plt.xlabel(column_name, fontsize=12)
plt.ylabel('Tần số', fontsize=12)

# Thêm lưới
plt.grid(True, alpha=0.3)

# Hiển thị biểu đồ
plt.show()
```

```python
# Tìm giá trị ngoại lai theo chỉ số PAPI tổng hợp (không có trọng số)
# Tính Q1, Q3 và IQR
Q1 = df[column_name].quantile(0.25)
Q3 = df[column_name].quantile(0.75)
IQR = Q3 - Q1

# Xác định ngưỡng ngoại lai
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Tìm các giá trị ngoại lai
outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]

# Vẽ boxplot để trực quan hóa
plt.figure(figsize=(8, 6))
sns.boxplot(x=df[column_name])
plt.title(f'Boxplot of {column_name}')
plt.xlabel(column_name)
```

```python
outlier_indices = df.index[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]
column_index = df.columns.get_loc(column_name)
df.iloc[outlier_indices, [0, column_index]]
```
