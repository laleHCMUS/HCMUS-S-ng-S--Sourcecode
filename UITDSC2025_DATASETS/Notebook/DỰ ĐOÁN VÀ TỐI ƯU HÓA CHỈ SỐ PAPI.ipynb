```python
# Các chỉ số để tạo lag/delta/pctchg
features = [
    "Chỉ số PAPI tổng hợp (không có trọng số)",
]

# Tạo biến động theo tỉnh
for col in features:
    df_all[f"{col}_lag1"] = df_all.groupby("Tỉnh/Thành phố")[col].shift(1)
    df_all[f"{col}_delta1"] = df_all[col] - df_all[f"{col}_lag1"]
    df_all[f"{col}_pctchg"] = df_all[f"{col}_delta1"] / df_all[f"{col}_lag1"]

# Kiểm tra lại cột year còn không
print("year" in df_all.columns)
```

```python
df_all["year_num"] = df_all["year"] - df_all["year"].min() # Chuẩn hóa năm thành số bắt đầu từ 0

# copy year sang year_raw để còn split train/val/test
df_all["year_raw"] = df_all["year"]

# Mỗi dòng sẽ có đúng 1 cột = 1, còn lại = 0 (One-hot encoding cột năm)
df_all = pd.get_dummies(df_all, columns=["year"], prefix="y")
```

```python
train_df = df_all[df_all["year_raw"].between(2019, 2022)]
val_df   = df_all[df_all["year_raw"] == 2023]
test_df  = df_all[df_all["year_raw"] == 2024]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error

# ---- 1. Xác định target ----
target = "Chỉ số PAPI tổng hợp (không có trọng số)" # => Bài toán hồi quy

# ---- 2. Xác định các cột cần loại bỏ ----
drop_cols = [col for col in df_all.columns if target in col] + ["year_raw"]

# ---- 3. One-hot encode cho cột dạng object (ví dụ Tỉnh/Thành phố)
def encode_df(df):
    df_enc = pd.get_dummies(df, columns=df.select_dtypes(include=["object"]).columns, drop_first=False)
    return df_enc

X_train = encode_df(train_df.drop(columns=drop_cols))
y_train = train_df[target]

X_val   = encode_df(val_df.drop(columns=drop_cols))
y_val   = val_df[target]

X_test  = encode_df(test_df.drop(columns=drop_cols))
y_test  = test_df[target]

# ---- 4. Align columns giữa train/val/test để tránh mismatch ----
X_val = X_val.reindex(columns=X_train.columns, fill_value=0)
X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

# ---- 5. Train model ----
model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(X_train, y_train)

# ---- 6. Đánh giá ----
y_val_pred  = model.predict(X_val)
y_test_pred = model.predict(X_test)

print("Validation:")
print("  R2   :", r2_score(y_val, y_val_pred))
print("  RMSE :", np.sqrt(mean_squared_error(y_val, y_val_pred)))

print("\nTest:")
print("  R2   :", r2_score(y_test, y_test_pred))
print("  RMSE :", np.sqrt(mean_squared_error(y_test, y_test_pred)))
```

```python
# ---- 6. Feature importance ----
importances = pd.Series(model.feature_importances_, index=X_train.columns)
importances = importances.sort_values(ascending=False)

plt.figure(figsize=(8, 5))
importances.head(15).plot(kind="barh")
plt.gca().invert_yaxis()
plt.title("Top 15 Feature Importances (RandomForest)")
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV
# ---- 1. Định nghĩa parameter grid ----
param_dist = {
    "n_estimators": [100, 200, 300, 500],
    "max_depth": [None, 5, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "max_features": ["auto", "sqrt", "log2"]
}

# ---- 2. Khởi tạo mô hình và RandomizedSearchCV ----
rf = RandomForestRegressor(random_state=42)

random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=20,            # số lần thử
    cv=3,                 # 3-fold cross-validation
    scoring="r2",         # tối ưu theo R²
    verbose=2,
    random_state=42,
    n_jobs=-1
)

# ---- 3. Train trên tập huấn luyện ----
random_search.fit(X_train, y_train)

# ---- 4. In kết quả ----
print("Best Params:", random_search.best_params_)
print("Best CV Score (R²):", random_search.best_score_)

# ---- 5. Đánh giá lại trên Val/Test ----
best_model = random_search.best_estimator_

y_val_pred  = best_model.predict(X_val)
y_test_pred = best_model.predict(X_test)

print("\nValidation:")
print("  R2   :", r2_score(y_val, y_val_pred))
print("  RMSE :", np.sqrt(mean_squared_error(y_val, y_val_pred)))

print("\nTest:")
print("  R2   :", r2_score(y_test, y_test_pred))
print("  RMSE :", np.sqrt(mean_squared_error(y_test, y_test_pred)))
```

```python
import optuna

# ----- 1. Định nghĩa objective -----
def objective(trial):
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 200, 1200),
        "max_depth": trial.suggest_int("max_depth", 3, 10),
        "learning_rate": trial.suggest_float("learning_rate", 0.01, 0.3, log=True),
        "subsample": trial.suggest_float("subsample", 0.5, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.5, 1.0),
        "min_child_weight": trial.suggest_int("min_child_weight", 1, 10),
        "reg_alpha": trial.suggest_float("reg_alpha", 1e-8, 10.0, log=True),
        "reg_lambda": trial.suggest_float("reg_lambda", 1e-8, 10.0, log=True),
        "random_state": 42,
        "n_jobs": -1,
        "objective": "reg:squarederror"
    }

    model = XGBRegressor(**params)

    model.fit(X_train, y_train)

    preds = model.predict(X_val)
    return r2_score(y_val, preds)

# ----- 2. Tạo study -----
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=40)

# ----- 3. Train lại với best params -----
best_params = study.best_params
print("Best Params:", best_params)

best_model = XGBRegressor(**best_params, objective="reg:squarederror", random_state=42, n_jobs=-1)
best_model.fit(X_train, y_train)

y_val_pred = best_model.predict(X_val)
y_test_pred = best_model.predict(X_test)

print("\nValidation R2:", r2_score(y_val, y_val_pred), "RMSE:", np.sqrt(mean_squared_error(y_val, y_val_pred)))
print("Test R2:", r2_score(y_test, y_test_pred), "RMSE:", np.sqrt(mean_squared_error(y_test, y_test_pred)))
```

```python
def optimize_papi_per_province_onehot(model, X_val, province_prefix='Tỉnh/Thành phố_', target='Chỉ số tổng hợp PAPI (không có trọng số)', top_n=5):
    from itertools import product

    # ---- 1. Lấy top_n features quan trọng ----
    importances = model.feature_importances_
    feat_names = X_val.drop(columns=[target]).columns
    sorted_idx = importances.argsort()[::-1]
    top_features = [feat_names[i] for i in sorted_idx[:top_n] if not feat_names[i].startswith(province_prefix)]

    # ---- 2. Lấy danh sách tỉnh từ one-hot ----
    province_cols = [c for c in X_val.columns if c.startswith(province_prefix)]
    provinces = [c.replace(province_prefix, '') for c in province_cols]

    best_values_dict = {}

    for prov_col, prov_name in zip(province_cols, provinces):
        df_prov = X_val[X_val[prov_col] == 1].copy()
        if df_prov.empty:
            continue

        baseline = df_prov.drop(columns=[target]).mean().to_dict()

        # ---- 3. Grid search top features ----
        feature_ranges = {}
        for feat in top_features:
            min_val = df_prov[feat].min()
            max_val = df_prov[feat].max()
            feature_ranges[feat] = np.linspace(min_val, max_val, 5)

        combinations = list(product(*feature_ranges.values()))
        best_pred = -np.inf
        best_comb = None

        for comb in combinations:
            x_test = baseline.copy()
            x_test.update(dict(zip(top_features, comb)))
            # giữ tỉnh hiện tại = 1, các tỉnh khác = 0
            for c in province_cols:
                x_test[c] = 1 if c == prov_col else 0
            x_df = pd.DataFrame([x_test])
            pred = model.predict(x_df)[0]
            if pred > best_pred:
                best_pred = pred
                best_comb = dict(zip(top_features, comb))

        # ---- 4. Lưu thêm PAPI_max ----
        best_values_dict[prov_name] = {**best_comb, "PAPI_max": best_pred}

    return best_values_dict
```

```python
# Ghép lại X_val + target để dùng trong hàm
df_val = X_val.copy()
df_val['Chỉ số tổng hợp PAPI (không có trọng số)'] = y_val
```
