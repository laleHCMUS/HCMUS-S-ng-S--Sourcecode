## Outliers check

```python
def show_outliers_for_cols(data, cols):
    results = {}
    for col in cols:
        q1 = data[col].quantile(0.25)
        q3 = data[col].quantile(0.75)
        iqr = q3 - q1
        results[col] = data[(data[col] < (q1 - 1.5 * iqr)) | (data[col] > (q3 + 1.5 * iqr))]
    return results
```

```python
cols = ['feature 2','feature 3','feature 5','feature 6']
outlier_dict = show_outliers_for_cols(provincial_2022_copy, cols)
```
