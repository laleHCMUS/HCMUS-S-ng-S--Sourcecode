```python
df.columns = df.loc["Tỉnh/Thành phố trực thuộc trung ương.1"]
# Xóa dòng đầu sau khi đã dùng làm tên cột
df = df.drop("Tỉnh/Thành phố trực thuộc trung ương.1")
# Reset index để đưa 'Tỉnh/Thành phố' thành cột
df = df.reset_index()
df = df.rename(columns={'index': 'Tỉnh/Thành phố trực thuộc trung ương.1'})
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
# Chọn các feature liên quan đến chuyển đổi số
features = ["Chỉ số nội dung 2: Công khai, minh bạch trong hoạch định chính sách",
            "Chỉ số nội dung 3: Trách nhiệm giải trình với người dân","Chỉ số nội dung 4: Kiểm soát tham nhũng trong khu vực công",
            "Chỉ số nội dung 5: Thủ tục hành chính công*","Chỉ số nội dung 6: Cung ứng dịch vụ công","Chỉ số nội dung 8: Quản trị điện tử"]

X = df[features]

# Chuẩn hóa dữ liệu
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# KMeans clustering với 3 nhóm
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
labels_1 = kmeans.fit_predict(X_scaled)
```

```python
# Đánh giá mô hình
from sklearn.metrics import silhouette_score
inertia = kmeans.inertia_
sil_score = silhouette_score(X_scaled, labels_1)

print(f"Inertia (WCSS): {inertia:.2f}")
print(f"Silhouette Score: {sil_score:.3f}")
```

```python
# Giảm chiều dữ liệu bằng PCA để cải thiện phân tách
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

kmeans = KMeans(n_clusters=3, random_state=42)
labels_2 = kmeans.fit_predict(X_pca)

inertia = kmeans.inertia_
sil_score = silhouette_score(X_scaled, labels_2)

print(f"Inertia (WCSS): {inertia:.2f}")
print(f"Silhouette Score: {sil_score:.3f}")
```

```python
# Thêm nhãn cluster vào DataFrame
df["Cluster"] = labels_2

# Chỉ lấy các cột số (loại bỏ tên tỉnh, v.v.)
numeric_cols = df.select_dtypes(include=["number"]).columns

# Tính trung bình theo cụm
cluster_summary = df.groupby("Cluster")[numeric_cols].mean()

print(cluster_summary)
```

```python
# Sắp xếp cụm theo điểm trung bình
cluster_order = cluster_summary.mean(axis=1).sort_values(ascending=False).index

# Tạo mapping: cụm có điểm cao nhất -> "Dẫn đầu", tiếp theo -> "Đang chuyển đổi", thấp nhất -> "Tụt hậu"
labels_order = ["Dẫn đầu", "Đang chuyển đổi", "Tụt hậu"]
cluster_map = {cluster_order[i]: labels_order[i] for i in range(3)}

# Gán nhãn
df["Muc_do"] = df["Cluster"].map(cluster_map)

# Xuất kết quả
print(df[["Tỉnh/Thành phố trực thuộc trung ương.1", "Muc_do"]])
```

```python
# Trực quan hóa kết quả phân cụm
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

df["PC1"] = X_pca[:, 0]
df["PC2"] = X_pca[:, 1]

# Vẽ scatter plot
plt.figure(figsize=(8,6))
colors = {"Đang chuyển đổi":"orange", "Dẫn đầu":"green", "Tụt hậu":"red"}

for label, color in colors.items():
    subset = df[df["Muc_do"] == label]
    plt.scatter(subset["PC1"], subset["PC2"], c=color, label=label, s=100, alpha=0.7, edgecolors="k")

# Gắn nhãn tên tỉnh (nếu muốn hiển thị)
for i, row in df.iterrows():
    plt.text(row["PC1"]+0.02, row["PC2"]+0.02, str(row["Tỉnh/Thành phố trực thuộc trung ương.1"]), fontsize=8)

plt.title("Phân cụm các tỉnh theo mức độ sẵn sàng chuyển đổi số (PCA 2D)")
plt.xlabel("Thành phần chính 1 (PC1)")
plt.ylabel("Thành phần chính 2 (PC2)")
plt.legend()
plt.grid(True)
plt.show()
```
