```python
df_2024 = pd.read_excel("PAPI2024.xlsx")
```

```python
df_2024
```

```python
df_2024 = df_2024.transpose()
df_2024
```

```python
df_2024 = df_2024.drop(["Tỉnh/Thành phố trực thuộc trung ương", "Thang điểm"])
```

```python
df_2024.columns = df_2024.loc["Tỉnh/Thành phố trực thuộc trung ương.1"]
# Xóa dòng đầu sau khi đã dùng làm tên cột
df_2024 = df_2024.drop("Tỉnh/Thành phố trực thuộc trung ương.1")
# Reset index để đưa 'Tỉnh/Thành phố' thành cột
df_2024 = df_2024.reset_index()
df_2024 = df_2024.rename(columns={'index': 'Tỉnh/Thành phố trực thuộc trung ương.1'})
```

```python
df_2024
```

```python
col = "Chỉ số PAPI tổng hợp (không có trọng số)"

# Sắp xếp theo cột
df_sorted_2024 = df_2024.sort_values(by=col).reset_index(drop=True)

# Chia thành 3 nhóm theo thứ tự
df_sorted_2024["Group"] = pd.qcut(df_sorted_2024[col], q=3, labels=["Tụt hậu", "Đang chuyển đổi", "Dẫn đầu"])

# Xem kết quả
print(df_sorted_2024[["Tỉnh/Thành phố trực thuộc trung ương.1", col, "Group"]])
```

```python
df_2019 = pd.read_excel("PAPI2019.xlsx")
```

```python
df_2019
```

```python
# Bỏ cột "Thang điểm"
df_2019 = df_2019.drop(columns=["Thang điểm"])
# Chuyển vị
df_2019 = df_2019.T
```

```python
df_2019.columns = df_2019.loc["Tỉnh/Thành phố trực thuộc trung ương"]
# Xóa dòng đầu sau khi đã dùng làm tên cột
df_2019 = df_2019.drop("Tỉnh/Thành phố trực thuộc trung ương")
# Reset index để đưa 'Tỉnh/Thành phố' thành cột
df_2019 = df_2019.reset_index()
df_2019 = df_2019.rename(columns={'index': 'Tỉnh/Thành phố trực thuộc trung ương'})
```

```python
df_2019
```

```python
col = "Chỉ số PAPI tổng hợp (không có trọng số)"

# Sắp xếp theo cột
df_sorted_2019 = df_2019.sort_values(by=col).reset_index(drop=True)

# Chia thành 3 nhóm theo thứ tự
df_sorted_2019["Group"] = pd.qcut(df_sorted_2019[col], q=3, labels=["Tụt hậu", "Đang chuyển đổi", "Dẫn đầu"])

# Xem kết quả
print(df_sorted_2019[["Tỉnh/Thành phố trực thuộc trung ương", col, "Group"]])
```

```python
df = df_2019[["Tỉnh/Thành phố trực thuộc trung ương"]].copy()

df_sorted_2020 = df_sorted_2020.rename(columns={"Tỉnh/Thành phố trực thuộc trung ương.1": "Tỉnh/Thành phố trực thuộc trung ương"})
df_sorted_2021 = df_sorted_2021.rename(columns={"Tỉnh/Thành phố trực thuộc trung ương.1": "Tỉnh/Thành phố trực thuộc trung ương"})
df_sorted_2022 = df_sorted_2022.rename(columns={"Tỉnh/Thành phố trực thuộc trung ương.1": "Tỉnh/Thành phố trực thuộc trung ương"})
df_sorted_2023 = df_sorted_2023.rename(columns={"Tỉnh/Thành phố trực thuộc trung ương.1": "Tỉnh/Thành phố trực thuộc trung ương"})
df_sorted_2024 = df_sorted_2024.rename(columns={"Tỉnh/Thành phố trực thuộc trung ương.1": "Tỉnh/Thành phố trực thuộc trung ương"})

df = df[["Tỉnh/Thành phố trực thuộc trung ương"]].merge(
    df_sorted_2019[["Tỉnh/Thành phố trực thuộc trung ương", "Group"]],
    on="Tỉnh/Thành phố trực thuộc trung ương",
    how="left"
)
df = df.rename(columns={"Group": "2019"})

df = df[["Tỉnh/Thành phố trực thuộc trung ương","2019"]].merge(
    df_sorted_2020[["Tỉnh/Thành phố trực thuộc trung ương", "Group"]],
    on="Tỉnh/Thành phố trực thuộc trung ương",
    how="left"
)
df = df.rename(columns={"Group": "2020"})

df = df[["Tỉnh/Thành phố trực thuộc trung ương","2019","2020"]].merge(
    df_sorted_2021[["Tỉnh/Thành phố trực thuộc trung ương", "Group"]],
    on="Tỉnh/Thành phố trực thuộc trung ương",
    how="left"
)
df = df.rename(columns={"Group": "2021"})

df = df[["Tỉnh/Thành phố trực thuộc trung ương","2019","2020","2021"]].merge(
    df_sorted_2022[["Tỉnh/Thành phố trực thuộc trung ương", "Group"]],
    on="Tỉnh/Thành phố trực thuộc trung ương",
    how="left"
)
df = df.rename(columns={"Group": "2022"})

df = df[["Tỉnh/Thành phố trực thuộc trung ương","2019","2020","2021","2022"]].merge(
    df_sorted_2023[["Tỉnh/Thành phố trực thuộc trung ương", "Group"]],
    on="Tỉnh/Thành phố trực thuộc trung ương",
    how="left"
)
df = df.rename(columns={"Group": "2023"})

df = df[["Tỉnh/Thành phố trực thuộc trung ương","2019","2020","2021","2022","2023"]].merge(
    df_sorted_2024[["Tỉnh/Thành phố trực thuộc trung ương", "Group"]],
    on="Tỉnh/Thành phố trực thuộc trung ương",
    how="left"
)
df = df.rename(columns={"Group": "2024"})

print(df)
```
